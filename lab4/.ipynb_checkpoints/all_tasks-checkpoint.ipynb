```python
# Импорт библиотек для работы с данными и графиками
import csv    # для чтения CSV файлов
import math   # для математических операций
import statistics  # для статистических вычислений
import matplotlib.pyplot as plt  # для построения графиков
from datetime import datetime  # для работы с датой и временем
```

```python
# Загрузка данных о продажах кофе из CSV файла
def load_coffee_data(filename):
    """
    Загружает данные из CSV файла и возвращает список словарей
    Каждая строка файла становится отдельным словарем с данными
    """
    data_records = []  # создаем пустой список для хранения данных
    with open(filename, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)  # создаем reader для чтения как словарей
        for record in reader:  # проходим по каждой строке файла
            data_records.append(record)  # добавляем строку в список
    return data_records

# Загружаем данные из файла
coffee_records = load_coffee_data('Coffe_sales.csv')
print(f"Успешно загружено записей: {len(coffee_records)}")

feature_names = list(coffee_records[0].keys())  # получаем названия всех признаков
num_features = len(feature_names)  # считаем количество признаков
print(f"Количество признаков: {num_features}")
print("Названия полей в данных:", feature_names)
```

```python
# Преобразование ВСЕХ признаков в числовые значения
def convert_all_features_to_numeric(data):
    """
    Преобразовывает все текстовые и категориальные данные в числа
    чтобы компьютер мог их анализировать математическими методами
    """
    # Создаем список для хранения преобразованных данных
    # Каждый элемент этого списка будет представлять одну запись с числовыми значениями
    converted_data = []
    
    # Создаем таблицы перевода текстовых значений в числа
    # Для способов оплаты: наличные=0, карта=1, мобильный=2
    cash_type_mapping = {'cash': 0, 'card': 1, 'mobile': 2}
    
    # Собираем все уникальные названия кофе из данных
    # Множество автоматически удаляет дубликаты
    coffee_names_set = set()
    
    # Проходим по всем записям данных чтобы собрать все виды кофе
    for record in data:
        # Из каждой записи извлекаем название кофе
        coffee_name = record['coffee_name']
        # Добавляем в множество (дубликаты игнорируются)
        coffee_names_set.add(coffee_name)
    
    # Преобразуем множество в отсортированный список для порядка
    coffee_names_list = sorted(list(coffee_names_set))
    
    # Создаем таблицу перевода названий кофе в числа
    coffee_mapping = {}
    # Присваиваем каждому виду кофе уникальный номер начиная с 0
    number = 0
    for coffee_name in coffee_names_list:
        coffee_mapping[coffee_name] = number
        number += 1
    
    # Таблица перевода для времени суток (учитывает регистр)
    time_mapping = {
        'morning': 0, 'afternoon': 1, 'night': 2,
        'Morning': 0, 'Afternoon': 1, 'Night': 2
    }
    
    # Таблица перевода для дней недели (сокращения + полные названия)
    weekday_mapping = {
        # Полные названия
        'monday': 0, 'tuesday': 1, 'wednesday': 2, 'thursday': 3, 
        'friday': 4, 'saturday': 5, 'sunday': 6,
        # Сокращения
        'mon': 0, 'tue': 1, 'wed': 2, 'thu': 3, 'fri': 4, 'sat': 5, 'sun': 6,
        # Сокращения с большой буквы
        'Mon': 0, 'Tue': 1, 'Wed': 2, 'Thu': 3, 'Fri': 4, 'Sat': 5, 'Sun': 6
    }
    
    # Таблица перевода для месяцев (3-буквенные сокращения)
    month_mapping = {
        # Полные названия
        'january': 0, 'february': 1, 'march': 2, 'april': 3, 'may': 4, 'june': 5,
        'july': 6, 'august': 7, 'september': 8, 'october': 9, 'november': 10, 'december': 11,
        # Сокращения
        'jan': 0, 'feb': 1, 'mar': 2, 'apr': 3, 'may': 4, 'jun': 5,
        'jul': 6, 'aug': 7, 'sep': 8, 'oct': 9, 'nov': 10, 'dec': 11,
        # Сокращения с большой буквы
        'Jan': 0, 'Feb': 1, 'Mar': 2, 'Apr': 3, 'May': 4, 'Jun': 5,
        'Jul': 6, 'Aug': 7, 'Sep': 8, 'Oct': 9, 'Nov': 10, 'Dec': 11
    }
    
    # Основной цикл: преобразуем каждую запись данных
    for record in data:
        # Создаем новую запись для хранения числовых значений
        numeric_record = {}
        
        # Преобразуем уже числовые признаки из строк в числа с плавающей точкой
        numeric_record['hour_of_day'] = float(record['hour_of_day'])
        numeric_record['money'] = float(record['money'])
        numeric_record['Weekdaysort'] = float(record['Weekdaysort'])
        numeric_record['Monthsort'] = float(record['Monthsort'])
        
        # Преобразуем способ оплаты из текста в число используя таблицу перевода
        cash_type = record['cash_type'].lower()
        if cash_type in cash_type_mapping:
            numeric_record['cash_type'] = cash_type_mapping[cash_type]
        else:
            numeric_record['cash_type'] = 0
        
        # Преобразуем название кофе из текста в число используя таблицу перевода
        coffee_name = record['coffee_name']
        if coffee_name in coffee_mapping:
            numeric_record['coffee_name'] = coffee_mapping[coffee_name]
        else:
            numeric_record['coffee_name'] = 0
        
        # Преобразуем время суток 
        time_of_day = record['Time_of_Day']  
        if time_of_day in time_mapping:
            numeric_record['Time_of_Day'] = time_mapping[time_of_day]
        else:
            numeric_record['Time_of_Day'] = 0
        
        # Преобразуем день недели 
        weekday = record['Weekday'] 
        if weekday in weekday_mapping:
            numeric_record['Weekday'] = weekday_mapping[weekday]
        else:
            numeric_record['Weekday'] = 0
        
        # Преобразуем название месяца 
        month_name = record['Month_name'] 
        if month_name in month_mapping:
            numeric_record['Month_name'] = month_mapping[month_name]
        else:
            numeric_record['Month_name'] = 0
        
        # Обрабатываем дату: извлекаем числовые компоненты (день, месяц, год)
        try:
            # Парсим строку даты в объект datetime
            date_obj = datetime.strptime(record['Date'], '%Y-%m-%d')
            # Извлекаем день месяца как число
            numeric_record['day_of_month'] = float(date_obj.day)
            # Извлекаем день недели (0=понедельник, 6=воскресенье)
            numeric_record['day_of_week'] = float(date_obj.weekday())
            # Извлекаем номер месяца
            numeric_record['month'] = float(date_obj.month)
            # Извлекаем год
            numeric_record['year'] = float(date_obj.year)
            # Определяем является ли день выходным (суббота или воскресенье)
            if date_obj.weekday() >= 5:
                numeric_record['is_weekend'] = 1.0
            else:
                numeric_record['is_weekend'] = 0.0
        except:
            # Если возникла ошибка при парсинге даты, используем значения по умолчанию
            numeric_record['day_of_month'] = 0.0
            numeric_record['day_of_week'] = 0.0
            numeric_record['month'] = 0.0
            numeric_record['year'] = 0.0
            numeric_record['is_weekend'] = 0.0
        
        # Обработка времени (учитывает миллисекунды)
        try:
            time_str = record['Time']
            # Пробуем разные форматы времени
            try:
                # Формат: 10:15:50:520000 (часы:минуты:секунды:миллисекунды)
                parts = time_str.split(':')
                if len(parts) >= 3:
                    hours = int(parts[0])
                    minutes = int(parts[1])
                    seconds = int(parts[2])
                    
                    numeric_record['hour'] = float(hours)
                    numeric_record['minute'] = float(minutes) 
                    numeric_record['second'] = float(seconds)
                    
                    # Определяем является ли время часами пик (утро 8-10 или вечер 17-19)
                    if (8 <= hours <= 10) or (17 <= hours <= 19):
                        numeric_record['is_rush_hour'] = 1.0
                    else:
                        numeric_record['is_rush_hour'] = 0.0
                else:
                    raise ValueError("Неверный формат времени")
                    
            except:
                # Альтернативный парсинг (если не сработал первый способ)
                time_obj = datetime.strptime(time_str.split('.')[0], '%H:%M:%S')  # игнорируем миллисекунды
                numeric_record['hour'] = float(time_obj.hour)
                numeric_record['minute'] = float(time_obj.minute)
                numeric_record['second'] = float(time_obj.second)
                
                # Часы пик
                hour_value = time_obj.hour
                if (8 <= hour_value <= 10) or (17 <= hour_value <= 19):
                    numeric_record['is_rush_hour'] = 1.0
                else:
                    numeric_record['is_rush_hour'] = 0.0
                    
        except:
            # Если возникла ошибка при парсинге времени, используем значения по умолчанию
            numeric_record['hour'] = 0.0
            numeric_record['minute'] = 0.0
            numeric_record['second'] = 0.0
            numeric_record['is_rush_hour'] = 0.0
        
        # Добавляем преобразованную запись в результирующий список
        converted_data.append(numeric_record)
    
    # Выводим статистику о проделанной работе
    print(f"Преобразовано записей: {len(converted_data)}")
    print(f"Количество признаков в каждой записи: {len(converted_data[0])}")
    print(f"Найдено видов кофе: {len(coffee_mapping)}")
    
    # Возвращаем полностью числовой датасет
    return converted_data

# Вызываем функцию преобразования для исходных данных
all_numeric_data = convert_all_features_to_numeric(coffee_records)
```

```python
# Получение списка ВСЕХ числовых признаков
def get_all_numeric_columns(data):
    """
    Получает список всех числовых признаков из преобразованных данных
    """
    return list(data[0].keys())  # берем ключи из первой записи

# Получаем все числовые признаки
all_numeric_fields = get_all_numeric_columns(all_numeric_data)
print(f"Всего числовых признаков: {len(all_numeric_fields)}")
print("Список всех признаков:")
for i, field in enumerate(all_numeric_fields, 1):
    print(f"  {i:2d}. {field}")
```

```python
# Создание полной числовой матрицы из ВСЕХ признаков
def create_complete_numeric_matrix(data, numeric_columns):
    """
    Создает матрицу (список списков) содержащую ВСЕ числовые данные
    Каждая строка матрицы - одна запись, каждый столбец - один признак
    """
    numeric_matrix = []  # создаем пустую матрицу
    for record in data:  # проходим по каждой записи
        numeric_row = []  # создаем пустую строку матрицы
        for column in numeric_columns:  # для каждого числового столбца
            numeric_row.append(record[column])  # добавляем значение в строку
        numeric_matrix.append(numeric_row)  # добавляем строку в матрицу
    
    return numeric_matrix

# Создаем полную числовую матрицу
complete_data_matrix = create_complete_numeric_matrix(all_numeric_data, all_numeric_fields)
print(f"Создана полная числовая матрица:")
print(f"  Количество строк (записей): {len(complete_data_matrix)}")
print(f"  Количество столбцов (признаков): {len(complete_data_matrix[0])}")
print(f"  Общий объем данных: {len(complete_data_matrix)} × {len(complete_data_matrix[0])} = {len(complete_data_matrix) * len(complete_data_matrix[0])} значений")

print("\nПервые 3 строки матрицы (первые 6 признаков):")
for i in range(3):
    first_six = complete_data_matrix[i][:6]  # берем первые 6 признаков для краткости
    print(f"  Строка {i+1}: {[f'{x:.1f}' for x in first_six]}...")
```

```python
# Нормировка данных методом Min-Max Scaling для ВСЕХ признаков
def normalize_data_minmax(matrix):
    """
    Применяет Min-Max нормировку ко ВСЕМ признакам данных
    Преобразует все значения в диапазон от 0 до 1
    Формула: (x - min) / (max - min)
    """
    print("Применяем нормировку ко всем признакам...")
    normalized_matrix = []  # матрица для нормированных данных
    num_columns = len(matrix[0])  # количество столбцов (признаков)
    
    # Списки для хранения минимальных и максимальных значений каждого столбца
    min_values = []  # минимальные значения каждого признака
    max_values = []  # максимальные значения каждого признака
    
    # Находим минимумы и максимумы для каждого столбца (признака)
    for col_idx in range(num_columns):
        column_data = []  # создаем пустой список для значений столбца (признака)
        for row in matrix:  # проходим по каждой строке матрицы
            value = row[col_idx]  # берем значение из текущего столбца
            column_data.append(value)  # добавляем значение в список
        min_values.append(min(column_data))  # находим минимум
        max_values.append(max(column_data))  # находим максимум
    
    # Применяем нормировку ко всем данным
    for row in matrix:
        normalized_row = []  # новая нормированная строка
        for col_idx in range(num_columns):
            value = row[col_idx]  # исходное значение
            min_val = min_values[col_idx]  # минимум для этого столбца
            max_val = max_values[col_idx]  # максимум для этого столбца
            
            # Проверяем, чтобы не делить на ноль
            if max_val - min_val == 0:
                normalized_value = 0.0  # если все значения одинаковые
            else:
                # применяем формулу нормировки
                normalized_value = (value - min_val) / (max_val - min_val)
            
            normalized_row.append(normalized_value)  # добавляем нормированное значение
        normalized_matrix.append(normalized_row)  # добавляем строку в матрицу
    
    print("Нормировка завершена!")
    return normalized_matrix

# Применяем нормировку ко ВСЕМ данным
normalized_data = normalize_data_minmax(complete_data_matrix)
print("Данные после нормировки (первые 3 строки, первые 6 признаков):")
for i in range(3):
    first_six = normalized_data[i][:6]  # берем первые 6 признаков для краткости
    formatted_values = []
    for x in first_six:
        formatted_value = f"{x:.4f}"  # форматируем число до 4 знаков после запятой
        formatted_values.append(formatted_value)
    print(f"  Строка {i+1}: {formatted_values}...")
```

```python
# Функция для вычисления корреляционной матрицы
def compute_correlation_matrix(matrix):
    """
    Вычисляет корреляционную матрицу Пирсона для данных
    Корреляция показывает линейную зависимость между признаками
    """
    num_features = len(matrix[0])  # количество признаков
    num_samples = len(matrix)      # количество образцов (записей)
    corr_matrix = []  # создаем пустую корреляционную матрицу
    
    # Шаг 1: Вычисляем средние значения для каждого признака
    feature_means = []  # список для средних значений
    for j in range(num_features):
        feature_sum = 0.0  # создаем переменную для накопления суммы
        for row in matrix:  # проходим по каждой строке в матрице
            feature_sum += row[j]  # добавляем значение j-го признака из текущей строки к общей сумме
        feature_means.append(feature_sum / num_samples)  # среднее значение
    
    # Шаг 2: Вычисляем корреляции между всеми парами признаков
    # i - номер первого признака в паре
    for i in range(num_features):
        corr_row = []  # строка корреляционной матрицы
        # j - номер второго признака в паре
        for j in range(num_features):
            if i == j:
                # Корреляция признака с самим собой всегда равна 1
                corr_row.append(1.0)
            else:
                # Вычисляем ковариацию и стандартные отклонения
                covariance = 0.0  # ковариация (мера совместной изменчивости двух признаков)
                std_i = 0.0       # стандартное отклонение(мера разброса значений) i-го признака
                std_j = 0.0       # стандартное отклонение(мера разброса значений) j-го признака
                
                # Проходим по всем образцам (записям)
                for k in range(num_samples):
                    # Отклонения от средних
                    dev_i = matrix[k][i] - feature_means[i] # насколько значение i-го признака отклоняется от своего среднего
                    dev_j = matrix[k][j] - feature_means[j] # насколько значение j-го признака отклоняется от своего среднего
                    
                    # Накопление сумм для вычислений
                    covariance += dev_i * dev_j # насколько признаки i и j изменяются вместе
                    std_i += dev_i * dev_i # насколько вообще изменяется признак i
                    std_j += dev_j * dev_j # насколько вообще изменяется признак j
                
                # Вычисляем ковариацию и стандартные отклонения
                covariance /= num_samples # средняя ковариация (делим сумму на кол-во записей)
                std_i = math.sqrt(std_i / num_samples) # стандартное отклонение из дисперсии
                std_j = math.sqrt(std_j / num_samples) # стандартное отклонение из дисперсии
                
                # Проверяем деление на ноль
                # Если один из признаков имеет нулевое стандартное отклонение, то есть все знаечния одинаковые, то
                if std_i * std_j == 0:
                    correlation = 0.0
                else:
                    # Формула корреляции Пирсона
                    correlation = covariance / (std_i * std_j)
                
                corr_row.append(correlation)  # добавляем корреляцию в строку
        
        corr_matrix.append(corr_row)  # добавляем строку в матрицу
    
    return corr_matrix

# Вычисляем первую корреляционную матрицу (после нормировки)
corr_matrix_1 = compute_correlation_matrix(normalized_data)

print("=" * 70)
print("ПЕРВАЯ КОРРЕЛЯЦИОННАЯ МАТРИЦА")
print("(после нормировки всех признаков)")
print("=" * 70)

print(f"Размер матрицы: {len(corr_matrix_1)}x{len(corr_matrix_1[0])}")

# Покажем только первые 6 признаков для читаемости
print("\nПервые 6 признаков (для наглядности):")
header = " " * 20
for name in all_numeric_fields[:6]:
    header += f"{name:>15}"
print(header)

for i in range(6):
    row_str = f"{all_numeric_fields[i]:<20}"
    for j in range(6):
        row_str += f"{corr_matrix_1[i][j]:>15.4f}"
    print(row_str)

print("\nВЫВОД: Первая матрица показывает исходные взаимосвязи между всеми признаками")
print("       после нормировки. Значения близкие к 1 или -1 указывают на сильную")
print("       корреляцию, близкие к 0 - на слабую связь.")
```

```python
# Визуализируем корреляционную матрицу 
def plot_correlation_matrix(corr_matrix, feature_names, title):
    """
    Рисует тепловую карту корреляционной матрицы
    """
    # Создаем рисунок
    plt.figure(figsize=(12, 10))
    
    # Создаем тепловую карту
    heatmap = plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
    
    # Добавляем цветовую шкалу
    plt.colorbar(heatmap, label='Корреляция')
    
    # Настраиваем подписи осей
    plt.xticks(range(len(feature_names)), feature_names, rotation=45, ha='right')
    plt.yticks(range(len(feature_names)), feature_names)
    
    # Добавляем заголовок
    plt.title(title)
    
    # Добавляем значения корреляций в ячейки
    for i in range(len(corr_matrix)):
        for j in range(len(corr_matrix[i])):
            # Белый текст для темных ячеек, черный - для светлых
            text_color = 'white' if abs(corr_matrix[i][j]) > 0.5 else 'black'
            plt.text(j, i, f'{corr_matrix[i][j]:.2f}', 
                    ha='center', va='center', color=text_color, fontsize=8)
    
    # Настраиваем внешний вид
    plt.tight_layout()
    plt.show()

# Визуализируем первую корреляционную матрицу
print("ВИЗУАЛИЗАЦИЯ ПЕРВОЙ КОРРЕЛЯЦИОННОЙ МАТРИЦЫ")
print("(после нормировки данных)")
print("Цвета: синий = отрицательная корреляция, красный = положительная")

plot_correlation_matrix(corr_matrix_1, all_numeric_fields, 
                       "Корреляционная матрица после нормировки")

print("\nАНАЛИЗ РЕЗУЛЬТАТОВ:")
print("• Красные ячейки показывают признаки, которые растут вместе")
print("• Синие ячейки показывают признаки, которые изменяются в противоположных направлениях")  
print("• Белые/светлые ячейки показывают слабую связь между признаками")
print("• Диагональ всегда красная (каждый признак идеально коррелирует сам с собой)")
```

```python
# Поиск и анализ выбросов в данных 
def find_and_analyze_outliers(matrix, feature_names):
    """
    Анализирует данные на наличие выбросов методом межквартильного размаха (IQR)
    Выбросы - значения, которые значительно отличаются от большинства данных
    """
    print("\n" + "=" * 70)
    print("АНАЛИЗ ВЫБРОСОВ В ДАННЫХ")
    print("=" * 70)
    
    # Создаем копию матрицы для безопасной работы
    cleaned_matrix = []
    for row in matrix:
        cleaned_matrix.append(row.copy())  # copy() чтобы не испортить исходные данные
    
    num_features = len(feature_names)  # количество признаков
    outlier_indices = set()  # множество индексов строк с выбросами
    
    # Статистика по выбросам
    outlier_summary = []  # список для сводки по выбросам
    
    # ИНИЦИАЛИЗИРУЕМ ПЕРЕМЕННЫЕ ДЛЯ СТАТИСТИКИ
    total_outliers = 0
    features_with_outliers = 0
    
    # Анализируем каждый признак отдельно
    for feature_idx in range(num_features):
        feature_values = []  # создаем пустой список для значений признака
        for row in matrix:   # проходим по каждой строке в матрице
            value = row[feature_idx]  # берем значение нужного признака из текущей строки
            feature_values.append(value)  # добавляем значение в список
        
        # Сортируем значения для нахождения квартилей
        sorted_vals = sorted(feature_values)
        n = len(sorted_vals)  # количество значений
        
        # Находим первый квартиль 
        q1_index = n // 4 # позиция первого квартиля (25%)
        q1 = sorted_vals[q1_index] # значение первого квартиля
        
        # Находим третий квартиль 
        q3_index = 3 * n // 4 # позиция третьего квартиля (75%)  
        q3 = sorted_vals[q3_index] # значение третьего квартиля
        
        # Вычисляем межквартильный размах (IQR)
        iqr = q3 - q1
        
        # ОПРЕДЕЛЯЕМ ГРАНИЦЫ С ПРОВЕРКОЙ НА IQR = 0
        if iqr == 0:
            # Если IQR нулевой, используем расширенные границы [min, max]
            # чтобы не считать нормальные значения выбросами
            lower_bound = min(feature_values)
            upper_bound = max(feature_values)
            method_used = "min-max (IQR=0)"
        else:
            # Стандартный метод IQR
            lower_bound = q1 - 1.5 * iqr
            upper_bound = q3 + 1.5 * iqr
            method_used = "IQR"
        
        # Ищем выбросы для этого признака
        feature_outliers = []  # список выбросов для этого признака
        outlier_values = []  # значения выбросов
        
        for row_idx, value in enumerate(feature_values):
            if value < lower_bound or value > upper_bound:
                feature_outliers.append(row_idx) # запоминаем номер строки с выбросом
                outlier_values.append(value) # запоминаем значение выброса
                outlier_indices.add(row_idx) # добавляем в множество выбросов
        
        # СЧИТАЕМ СТАТИСТИКУ ВЫБРОСОВ
        outlier_count = len(feature_outliers)
        if outlier_count > 0:
            features_with_outliers += 1
            total_outliers += outlier_count
        
        # Заменяем выбросы граничными значениями
        for row in cleaned_matrix:
            value = row[feature_idx]
            if value < lower_bound:
                row[feature_idx] = lower_bound  # заменяем на нижнюю границу
            elif value > upper_bound:
                row[feature_idx] = upper_bound  # заменяем на верхнюю границу
        
        # Сохраняем статистику для этого признака
        outlier_summary.append({
            'feature': feature_names[feature_idx],
            'outlier_count': outlier_count,
            'outlier_percentage': (outlier_count / len(matrix)) * 100,
            'min_value': min(feature_values),
            'max_value': max(feature_values),
            'bounds': [lower_bound, upper_bound],
            'method': method_used  # добавляем информацию о методе
        })

    # Проверяем есть ли выбросы
    has_any_outliers = (total_outliers > 0)
    
    # Если выбросов нет, то сообщаем об этом
    if not has_any_outliers:
        print("ВЫБРОСОВ НЕ ОБНАРУЖЕНО")
        print("Все данные находятся в пределах статистических границ")
    else:
        # Выводим сводку по выбросам
        print("\nСВОДКА ПО ВЫБРОСАМ:")
        print("-" * 90)
        print(f"{'Признак':<20} {'Выбросы':<10} {'%':<8} {'Мин':<8} {'Макс':<8} {'Границы':<20} {'Метод'}")
        print("-" * 90)
    
        for summary in outlier_summary:
            if summary['outlier_count'] > 0:
                # Форматируем вывод границ
                bounds_str = f"[{summary['bounds'][0]:.3f}, {summary['bounds'][1]:.3f}]"
                print(f"{summary['feature']:<20} {summary['outlier_count']:<10} {summary['outlier_percentage']:<7.1f} "
                      f"{summary['min_value']:<7.3f} {summary['max_value']:<7.3f} {bounds_str:<20} {summary['method']}")
    
    # Выводим признаки без выбросов (только если их немного)
    no_outliers_features = []  # создаем пустой список
    for s in outlier_summary:  # проходим по всей статистике
        if s['outlier_count'] == 0:  # если выбросов нет
            no_outliers_features.append(s)  # добавляем в список
    
    if len(no_outliers_features) <= 10:  # если мало признаков без выбросов - покажем все
        print(f"\nПризнаки БЕЗ выбросов ({len(no_outliers_features)}):")
        for summary in no_outliers_features:
            bounds_str = f"[{summary['bounds'][0]:.3f}, {summary['bounds'][1]:.3f}]"
            print(f"  {summary['feature']:<18} - границы: {bounds_str} ({summary['method']})")
    else:
        # Если много признаков без выбросов - покажем только первые 5
        print(f"\nПризнаки БЕЗ выбросов (первые 5 из {len(no_outliers_features)}):")
        for summary in no_outliers_features[:5]:
            bounds_str = f"[{summary['bounds'][0]:.3f}, {summary['bounds'][1]:.3f}]"
            print(f"  {summary['feature']:<18} - границы: {bounds_str} ({summary['method']})")
        print(f"  ... и еще {len(no_outliers_features) - 5} признаков")
    
    # Выводим итоговую статистику
    print(f"\n" + "=" * 60)
    print("ИТОГОВАЯ СТАТИСТИКА:")
    print(f"Всего признаков: {num_features}")
    print(f"Признаков с выбросами: {features_with_outliers}")
    print(f"Всего обнаружено выбросов: {total_outliers}")
    print(f"Уникальных строк с выбросами: {len(outlier_indices)}")
    print(f"Записей до обработки: {len(matrix)}")
    print(f"Записей после обработки: {len(cleaned_matrix)}")
    print(f"Удалено записей: {len(matrix) - len(cleaned_matrix)}")
    print(f"Сохранено данных: {(len(cleaned_matrix)/len(matrix))*100:.1f}%")
    
    return cleaned_matrix

# Применяем функцию поиска и обработки выбросов
cleaned_data = find_and_analyze_outliers(normalized_data, all_numeric_fields)
```

```python
# Удаление тождественных экземпляров
def remove_duplicate_instances(matrix):
    """
    Удаляет полностью одинаковые (тождественные) экземпляры из данных
    Тождественные экземпляры - строки с одинаковыми значениями всех признаков
    """
    print("\n" + "=" * 70)
    print("УДАЛЕНИЕ ТОЖДЕСТВЕННЫХ ЭКЗЕМПЛЯРОВ")
    print("=" * 70)
    
    unique_instances = []  # список для уникальных экземпляров
    seen_instances = set()  # множество для отслеживания уже встреченных экземпляров
    duplicate_count = 0  # счетчик удаленных дубликатов
    
    # Проходим по всем строкам матрицы
    for row in matrix:
        # Преобразуем строку в кортеж для хранения в множестве
        # Списки нельзя хранить в множестве (они изменяемые), а кортежи - можно.
        row_tuple = tuple(row)
        
        # Проверяем, встречался ли такой экземпляр ранее
        if row_tuple not in seen_instances:
            # Если не встречался - добавляем в уникальные
            seen_instances.add(row_tuple)
            unique_instances.append(row)
        else:
            # Если встречался - увеличиваем счетчик дубликатов
            duplicate_count += 1
    
    # Выводим статистику
    print(f"Удалено тождественных экземпляров: {duplicate_count}")
    print(f"Уникальных записей до удаления: {len(matrix)}")
    print(f"Уникальных записей после удаления: {len(unique_instances)}")
    print(f"Процент удаленных записей: {(duplicate_count/len(matrix))*100:.2f}%")
    
    return unique_instances

# Удаляем дубликаты из очищенных данных
final_cleaned_data = remove_duplicate_instances(cleaned_data)
```

```python
# Преобразование в целочисленную форму (обобщенная сериализация)
def generalize_serialization(matrix, feature_names):
    """
    Преобразует вещественные данные в целочисленные по методичке обобщенной сериализации
    Это обратимое преобразование, сохраняющее информацию о данных
    """
    print("\n" + "=" * 70)
    print("ПРЕОБРАЗОВАНИЕ В ЦЕЛОЧИСЛЕННУЮ ФОРМУ")
    print("(обобщенная сериализация)")
    print("=" * 70)
    
    num_features = len(matrix[0])  # количество признаков
    integer_matrix = []  # матрица для целочисленных данных
    powers = []  # список мощностей признаков
    
    # Заранее определяем дискретность для каждого признака (как в примере с длиной детали из методички)
    discretization_steps = {}
    
    for feature_idx, feature_name in enumerate(feature_names):
        # Определяем дискретность для каждого типа признаков
        if feature_name in ['hour_of_day', 'hour']:
            discretization_steps[feature_idx] = 1.0  # час - дискретность 1
        elif feature_name == 'money':
            discretization_steps[feature_idx] = 0.01  # деньги - 0.01
        elif feature_name in ['minute', 'second']:
            discretization_steps[feature_idx] = 1.0  # минуты, секунды - 1
        elif feature_name in ['cash_type', 'Time_of_Day', 'Weekday', 'Month_name', 
                            'coffee_name', 'is_weekend', 'is_rush_hour']:
            discretization_steps[feature_idx] = 1.0  # категориальные - 1.0
        else:
            discretization_steps[feature_idx] = 0.1  # остальные - 0.1
    
    # Вычисляем мощность для каждого признака
    for feature_idx in range(num_features):
        # Собираем все значения признака 
        feature_values = []  # создаем пустой список для значений признака
        for row in matrix:  # проходим по каждой строке матрицы
            value = row[feature_idx]  # берем значение из нужного столбца
            feature_values.append(value)  # добавляем значение в список
        
        # Находим минимальное и максимальное значения
        min_val = min(feature_values)
        max_val = max(feature_values)
        
        # Берем заранее определенную дискретность
        discretization_step = discretization_steps[feature_idx]
        
        # ВЫЧИСЛЯЕМ МОЩНОСТЬ по формуле из методички
        # мощность = (max - min) / дискретность + 1
        if discretization_step == 0:
            power = 1
        else:
            power = int((max_val - min_val) / discretization_step) + 1
        
        powers.append(power)  # сохраняем мощность
        print(f"{feature_names[feature_idx]}: дискретность={discretization_step}, мощность={power}")
    
    # Преобразуем данные в целочисленную форму по формуле из методички
    for row in matrix:
        integer_row = []  # целочисленная строка
        for feature_idx in range(num_features):
            value = row[feature_idx]  # исходное значение
            
            # Находим минимум для этого признака
            min_val = float('inf')  # начинаем с очень большого числа
            for r in matrix:  # проходим по всем строкам матрицы
                current_value = r[feature_idx]  # берем значение из текущего столбца
                if current_value < min_val:  # если нашли меньше текущего минимума
                    min_val = current_value  # обновляем минимум
            
            # Берем заранее определенную дискретность
            discretization_step = discretization_steps[feature_idx]
            
            # Применяем формулу из методички:
            # [(x - min) / дискретность]
            if discretization_step == 0:
                integer_value = 0
            else:
                integer_value = int((value - min_val) / discretization_step)
            
            integer_row.append(integer_value)  # добавляем целое значение
        integer_matrix.append(integer_row)  # добавляем строку в матрицу
    
    # Показываем пример преобразованных данных
    print("\nПервые 3 строки после преобразования в целочисленную форму:")
    for i in range(3):
        first_six = integer_matrix[i][:6]  # покажем первые 6 признаков
        print(f"Строка {i+1}: {first_six}...")
    
    return integer_matrix, powers

# Преобразуем очищенные данные в целочисленную форму
integer_data, powers = generalize_serialization(final_cleaned_data, all_numeric_fields)
```

```python
# Уменьшение размерности данных до 3 
def reduce_dimension_to_3(matrix, powers, feature_names):
    """
    Уменьшает размерность данных до 3 методом обобщенной сериализации
    Объединяет признаки с наименьшими мощностями для сохранения информации
    """
    print("\n" + "=" * 70)
    print("УМЕНЬШЕНИЕ РАЗМЕРНОСТИ ДАННЫХ ДО 3")
    print("=" * 70)
    
    # Создаем копии исходных данных чтобы не изменять оригинальные значения
    current_matrix = []
    for row in matrix:
        current_matrix.append(row.copy())
    
    current_powers = powers.copy()
    current_names = feature_names.copy()
    
    print(f"Начальная размерность: {len(current_powers)} признаков")
    print(f"Начальные мощности: {current_powers}")
    
    # Список для хранения информации о каждом шаге объединения
    steps_info = []
    step_count = 0
    
    # Основной цикл уменьшения размерности - работаем пока не получим 3 признака
    while len(current_powers) > 3:
        step_count += 1
        
        # Создаем список пар (индекс признака, его мощность) для сортировки
        power_index_pairs = []
        for i in range(len(current_powers)):
            power_index_pairs.append((i, current_powers[i]))
        
        # Применяем пузырьковую сортировку для нахождения признаков с наименьшими мощностями
        # Сортируем по возрастанию мощности (второй элемент кортежа)
        n = len(power_index_pairs)
        for i in range(n):
            for j in range(0, n - i - 1):
                # Сравниваем мощности соседних элементов
                if power_index_pairs[j][1] > power_index_pairs[j + 1][1]:
                    # Если текущий элемент больше следующего, меняем их местами
                    temp = power_index_pairs[j]
                    power_index_pairs[j] = power_index_pairs[j + 1]
                    power_index_pairs[j + 1] = temp
        
        # После сортировки первые два элемента имеют наименьшие мощности
        idx1, power1 = power_index_pairs[0]  # индекс и мощность первого признака для объединения
        idx2, power2 = power_index_pairs[1]  # индекс и мощность второго признака для объединения
        
        # Сохраняем информацию о текущем шаге для последующего вывода
        step_info = {
            'step': step_count,
            'name1': current_names[idx1],
            'name2': current_names[idx2], 
            'power1': power1,
            'power2': power2,
            'new_power': power1 * power2
        }
        steps_info.append(step_info)
        
        print(f"Шаг {step_count}: Объединяем '{current_names[idx1]}' и '{current_names[idx2]}'")
        print(f"  Мощности: {power1} × {power2} = {power1 * power2}")
        
        # Создаем новые структуры данных для хранения результатов объединения
        new_matrix = []    # матрица данных после объединения
        new_powers = []    # список новых мощностей признаков
        new_names = []     # список новых названий признаков
        
        # Переносим все признаки кроме двух объединяемых в новые списки
        # Сохраняем исходный порядок признаков
        for i in range(len(current_powers)):
            if i != idx1 and i != idx2:
                new_powers.append(current_powers[i])
                new_names.append(current_names[i])
        
        # Добавляем объединенный признак в конец списка
        # Мощность объединенного признака равна произведению мощностей исходных признаков
        new_power = power1 * power2
        new_powers.append(new_power)
        new_names.append(f"Объединение_{step_count}")
        
        # Формируем новую матрицу данных с объединенными признаками
        for row in current_matrix:
            new_row = []  # новая строка данных
            
            # Добавляем значения всех признаков кроме объединяемых
            for i in range(len(current_powers)):
                if i != idx1 and i != idx2:
                    new_row.append(row[i])
            
            # Вычисляем значение объединенного признака по формуле из методички:
            # z = y1 + W1 * y2, где W1 - мощность первого признака
            combined_value = row[idx1] + power1 * row[idx2]
            new_row.append(combined_value)
            
            # Добавляем новую строку в матрицу
            new_matrix.append(new_row)
        
        # Обновляем текущие данные для следующей итерации
        current_matrix = new_matrix
        current_powers = new_powers
        current_names = new_names
        
        print(f"  Новая размерность: {len(current_powers)} признаков")
    
    # Выводим подробную историю всех объединений
    print("\n" + "=" * 50)
    print("ПОДРОБНАЯ ИСТОРИЯ ОБЪЕДИНЕНИЙ:")
    print("-" * 50)
    
    for step in steps_info:
        print(f"Шаг {step['step']:2d}:")
        print(f"  Признак 1: {step['name1']}")
        print(f"    Мощность: {step['power1']}")
        print(f"  Признак 2: {step['name2']}")
        print(f"    Мощность: {step['power2']}")
        print(f"  Объединенный признак: мощность = {step['new_power']}")
        print()
    
    # Выводим финальные результаты
    print(f"\nФИНАЛЬНЫЕ РЕЗУЛЬТАТЫ:")
    print(f"Целевая размерность достигнута: {len(current_powers)} признака")
    print(f"Мощности финальных признаков: {current_powers}")
    print("Состав финальных компонент:")
    for i in range(len(current_names)):
        name = current_names[i]
        power = current_powers[i]
        print(f"  Компонента {i+1}: {name}")
        print(f"    Мощность: {power}")
    
    return current_matrix, current_powers, current_names

# Применяем функцию уменьшения размерности к целочисленным данным
reduced_data, reduced_powers, reduced_names = reduce_dimension_to_3(integer_data, powers, all_numeric_fields)
```

```python
# Визуализация третьей корреляционной матрицы (3x3)
def plot_small_correlation_matrix(corr_matrix, feature_names, title):
    """
    Рисует тепловую карту для маленькой корреляционной матрицы (3x3)
    Специально для матрицы после уменьшения размерности
    """
    # Создаем рисунок меньшего размера для 3 признаков
    plt.figure(figsize=(8, 6))
    
    # Создаем тепловую карту
    heatmap = plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
    
    # Добавляем цветовую шкалу
    plt.colorbar(heatmap, label='Корреляция')
    
    # Настраиваем подписи осей - для 3 признаков можно сделать покрупнее
    plt.xticks(range(len(feature_names)), feature_names, rotation=0, fontsize=12)
    plt.yticks(range(len(feature_names)), feature_names, fontsize=12)
    
    # Добавляем заголовок
    plt.title(title, fontsize=14, pad=20)
    
    # Добавляем значения корреляций в ячейки - делаем текст крупнее
    for i in range(len(corr_matrix)):
        for j in range(len(corr_matrix[i])):
            # Белый текст для темных ячеек, черный - для светлых
            text_color = 'white' if abs(corr_matrix[i][j]) > 0.5 else 'black'
            plt.text(j, i, f'{corr_matrix[i][j]:.3f}', 
                    ha='center', va='center', color=text_color, fontsize=14, fontweight='bold')
    
    # Настраиваем внешний вид
    plt.tight_layout()
    plt.show()

plot_small_correlation_matrix(corr_matrix_3, reduced_names,
                             "Корреляционная матрица после уменьшения размерности")
```
